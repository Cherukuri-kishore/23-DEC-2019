# 1.show the count of different types of species

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Iris.csv")
df.groupby(['Species']).size()
```

```python
df.describe()
```

# 3.scatterplot(for sepallength and petallength)

```python
plt.scatter(df.SepalLengthCm,df.PetalLengthCm)
```

# 4.bargraph

```python
plt.bar(df.SepalLengthCm,df.SepalWidthCm,df.PetalLengthCm,df.PetalWidthCm)
```

# jointplot(for sepallength and sepalwidth)

```python
sns.jointplot(df.SepalLengthCm,df.SepalWidthCm)
```

# facetgrid(for sepallength and sepalwidth for all species)

```python
g=sns.FacetGrid(df,col="SepalLengthCm")
g.map(plt.hist,"SepalWidthCm")
```

# boxplot(for species and sepalwidth)

```python
g = sns.boxplot(df.Species,df.SepalWidthCm)
```

# boxplot & stripplot(for sepallength and all species)

```python
sns.violinplot(df.PetalLengthCm,df.Species)
```
